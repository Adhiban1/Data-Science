### Random values

```python
import numpy as np
import pandas as pd

np.random.seed(0)
```

```python
n=100
df2 = pd.DataFrame(np.random.randint(0, 6, (1000, n)), columns=[f'Movie{i+1}' for i in range(n)])
df2
```

```python
df2_corr = df2.corr()
df2_corr
```

```python
def recommendation(movie, rating):
    sim = df2_corr[movie] * (rating - 2.5)
    # sim = sim.sort_values(ascending=False)
    return sim

recommendation('Movie10', 2).sort_values(ascending=False)
```

```python
movies = [f'Movie{i+1}' for i in range(n)]
sim = pd.DataFrame()
random.shuffle(movies)
ratings = [(movies[i], random.randint(0, 5)) for i in range(20)]

for movie, r in ratings:
    sim = pd.concat([sim, recommendation(movie, r)], axis=1)

sim.sum(axis=1).sort_values(ascending=False)
```
