# Time Series Forecast (TSF) | Linear regression

```python
import math
import random
from colorama import Fore
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
def convert(l, n):
    x = []
    y = []
    for i in range(len(l)-n+1):
        x.append(l[i:i+n-1])
        y.append(l[i+n-1])
    return x, y
```

# Prediction 1

```python
equation = 'x**3'
x1 = list(range(-20, 20))
y1 = eval(f'[{equation} for x in x1]')
plt.plot(y1)
plt.show()
print(f'y1:{Fore.CYAN}{y1[:10]}\b ...]')

x, y = convert(y1, 5)
for i, j in list(zip(x,y))[:5]:
    print(f'{Fore.BLUE}{i} {Fore.GREEN}{j}')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
model = LinearRegression()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
if len(y1) == 40:
    for _ in range(10):
        y1 = y1 + model.predict([y1[-4:]]).tolist()

plt.plot(x1, y1[:40])
plt.plot(range(19, 30), y1[40-1:], '--', label='prediction')
plt.legend()
plt.show()
```
