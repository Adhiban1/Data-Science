```python
import pandas as pd
```

```python
df = pd.read_csv('dataset\Salary_Dataset_with_Extra_Features.csv')
```

```python
df.head()
```

```python
temp = []
for i in df['Job Title']:
    if 'Python' in i:
        temp.append(True)
    else: temp.append(False)
df1 = df[temp]
```

```python
df2 = df1.groupby('Job Title').mean().sort_values('Salary', ascending=False)['Salary'].to_frame().reset_index()
df2.columns = ['Job Title', 'Average Salary']
df2 = df2.drop_duplicates()
df2
```

```python
df1 = df1.sort_values('Job Title')
```

```python
df1 = df1[['Rating', 'Company Name', 'Job Title', 'Salary',
       'Location', 'Employment Status']]
```

```python
df1.Salary = df1.Salary/100000
```

```python
df1.columns = ['Rating', 'Company Name', 'Job Title', 'Salary (L)/Year', 'Location',
       'Employment Status']
```

```python
df1.head()
```

```python
df1.to_csv('Python Jobs.csv', index=False)
```

```python
df2['Average Salary'] = df2['Average Salary']/100000
```

```python
df2.columns = ['Job Title', 'Average Salary (L)/Year']
```

```python
df2['Average Salary (L)/Year'] = df2['Average Salary (L)/Year'].round(2)
```

```python
df2.head()
```

```python
df2.to_csv('Python Average Salary.csv', index=False)
```
