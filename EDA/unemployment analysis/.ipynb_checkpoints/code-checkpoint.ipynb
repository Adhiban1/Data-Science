```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('unemployment analysis.csv')
df
```

```python
df1 = df.set_index('Country Code').drop('Country Name', axis=1).transpose()
df1
```

```python
fig = px.line(df1, markers=True)
fig.write_html('plot1.html')
```

```python
fig = px.line(df1[['IND', 'USA', 'GBR', 'MYS']], markers=True)
fig.show()
```

 - At covid time, Unemployment goes high

```python
for i in df1.columns:
    if i.startswith('U'):
        print(i)
```

```python
df2 = df1.mean().to_frame()
df2.columns = ['Value']
df2 = df2.sort_values('Value')
fig = px.bar(df2)
fig.show()
```
