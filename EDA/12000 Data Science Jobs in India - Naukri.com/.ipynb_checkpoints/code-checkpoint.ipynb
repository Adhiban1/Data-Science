```python
import pandas as pd
```

```python
df = pd.read_csv(r'12000 Data Science Jobs in India - Naukri.com\naukri_data_science_jobs_india.csv')
df.head()
```

```python
temp = []
for i in df['Skills/Description']:
    temp += i.split(', ')
temp = list(set(temp))
temp
```

```python
from rich.progress import track
```

```python
d = []
for j in track(temp, 'Skill'):
    k = 0
    for i in df['Skills/Description']:
        if  f' {j},' in i:
            k += 1
    d.append([j, k])

df1 = pd.DataFrame(d, columns=['Skill', 'Count'])
```

```python
df1 = df1.sort_values('Count', ascending=False)
```

```python
df1.head()
```

```python
df1.to_csv('Skill-count.csv', index=False)
```

```python
df1.shape
```

```python
s = 'Skills keywords for Data Science job description:\n' + ', '.join(df1.Skill)
with open('Skills Keywords.txt', 'w') as f:
    f.write(s)
```

```python
sheet = []
for i in track(df.Job_Role.unique()):
    a = df[df['Job_Role'] == i]
    temp = []
    for j in a['Skills/Description']:
        temp += j.split(', ')
    temp = list(set(temp))
    temp = ', '.join(temp)
    sheet.append([i, temp])
```

```python
sheet[0]
```

```python
df2 = pd.DataFrame(sheet, columns=['Job', 'Skills/Keywords'])
df2.head()
```

```python
df2.shape
```

```python
df2.to_csv('Jobs_skills_or_keywords.csv', index=False)
```

```python
sheet = {}
for i in track(df.Job_Role.unique()):
    a = df[df['Job_Role'] == i]
    temp = []
    for j in a['Skills/Description']:
        temp += j.split(', ')
    temp = list(set(temp))
    sheet[i] = temp

with open('Jobs_skills_or_keywords.json', 'w') as f:
    json.dump(sheet, f, indent=4)
```

```python
temp = []
for i in df.Location.unique():
    if 'Bengaluru' in i:
        i = i.replace('Bengaluru', 'Bangalore')
    temp += re.findall('[A-Za-z]+', i)
temp1 = list(set(temp))
sheet1 = []
for i in temp1:
    k = 0
    for j in temp:
        if j == i:
            k += 1
    sheet1.append([i, k])
df3 = pd.DataFrame(sheet1, columns=['Location', 'Count']).sort_values('Count', ascending=False)
df3.head(10)
```

```python
df3.to_csv('Location-count.csv', index=False)
```

```python
temp = []
for i in df.Location.unique():
    temp += i.split(', ')
temp1 = list(set(temp))
sheet2 = []
for i in temp1:
    sheet2.append([i, temp.count(i)])
df4 = pd.DataFrame(sheet2, columns=['Locations', 'Count']).sort_values('Count', ascending=False)
df4.head()
```

```python
df4.to_csv('Location-count(1).csv', index=False)
```

```python
temp3 = []
for i in df['Job Experience']:
    if re.findall('[A-Za-z]', i):
        temp3.append(False)
    else:
        temp3.append(True)
df6 = df[temp3]
df6.head()
```

```python
for i in df6['Job Experience'].unique():
    print(re.findall('[\d]+', i))
```

```python
sheet4 = []
for i in df6['Job Experience']:
    sheet4.append(re.findall('[\d]+', i))
df6[['Min Experience', 'Max Experience']] = pd.DataFrame(sheet4).astype(int)
df6.head()
```

```python
df6 = df6.drop('Job Experience', axis=1)
df6.head()
```

```python
df6.to_csv('cleaned_data.csv', index=False)
```

```python
df7 = df6.loc[df6['Min Experience'] == 0]
df7.head()
```

```python
temp4 = []
for i in df7.Job_Role:
    i = i.lower()
    if 'data' in i or 'scien' in i:
        temp4.append(True)
    else:
        temp4.append(False)
df7 = df7[temp4]
```

```python
df7.Location.unique()
```

```python
def search(*args):
    global df7
    k = 0
    for i in df7.Location:
        for j in args:
            if j in i.lower():
                k += 1
    return k
```

```python
search('bangalore', 'chennai')
```

```python
temp5 = []
for i in df7.Location:
    i = i.lower()
    if 'bangalore' in i or 'chennai' in i:
        temp5.append(True)
    else:
        temp5.append(False)
df7 = df7[temp5]
```

```python
temp6 = []

for i in df7['Skills/Description'].unique():
    temp6 += i.split(', ')
temp7 = list(set(temp6))
sheet5 = []
for i in temp7:
    k = 0
    for j in df7['Skills/Description']:
        if i in j:
            k += 1
    sheet5.append([i, k])
df8 = pd.DataFrame(sheet5, columns=['Keywords', 'Count']).sort_values('Count', ascending=False)
df8.head()
```
