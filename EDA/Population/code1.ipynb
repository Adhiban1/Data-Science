```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("world_population.csv")
df.head()
```

```python
df.columns
```

```python
df1 = df[['Country', '2022 Population',
       '2020 Population', '2015 Population', '2010 Population',
       '2000 Population', '1990 Population', '1980 Population',
       '1970 Population']]
df1 = df1.transpose()
df1
```

```python
df1 = df1.reset_index()
```

```python
df1
```

```python
list1 = df1.loc[0].to_list()
list1[0] = 'Year'
```

```python
df1.columns = list1
```

```python
df1
```

```python
df1 = df1.loc[1:]
df1
```

```python
df1 = df1[::-1]
df1
```

```python
fig = px.line(df1, x='Year', y=list1[1:], markers=True)
fig.write_html("population.html")
```
