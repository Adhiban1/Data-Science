```python
import pandas as pd
```

```python
df = pd.read_csv('jobs.csv', index_col=0)
```

```python
df.head()
```

```python
temp = []
for i in df['Job_title']:
    if 'Python' in i:
        temp.append(True)
    else: temp.append(False)
df1 = df[temp]
df1
```

```python
df2 = df1[['Job_title', 'skills_required']]
df2
```

```python
df2.to_csv('Python Job Skills.csv', index=False)
```
