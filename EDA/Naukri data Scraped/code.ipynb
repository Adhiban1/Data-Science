```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('IT_salaries.csv')
```

```python
df = df.dropna()
```

```python
df.head() # 1627 rows × 8 columns
```

```python
temp = []
for i in df.job_title:
    if 'Python' in i:
        temp.append(True)
    else: temp.append(False)
df1 = df[temp]
df1.head()
```

```python
import re
```

```python
def func1(a):
    a = a.replace(',', '')
    try:
        return round(int(re.findall('[\d]+', a)[0])/100000, 2)
    except:
        return np.nan
```

```python
def func2(a):
    a = a.replace(',', '')
    return round(int(re.findall('[\d]+', a)[1])/100000, 2)
```

```python
df1 = df1[df1.salary != 'Not Disclosed']
```

```python
df1['Min Salary'] = df1.salary.apply(func1)
```

```python
df1['Max Salary'] = df1.salary.apply(func2)
```

```python
df1.head()
```

```python
df1.experience.unique()
```

```python
def func3(x):
    return int(re.findall('[\d]+', x)[0])
def func4(x):
    try:
        return int(re.findall('[\d]+', x)[1])
    except:
        return 0


df1['Min experience'] = df1.experience.apply(func3)
df1['Max experience'] = df1.experience.apply(func4)
```

```python
df1.head()
```

```python
temp = []
for i in df1.location:
    if 'Chennai' in i or 'Hyderabad' in i or 'Bangalore' in i:
        temp.append(True)
    else: temp.append(False)
df1 = df1[temp]
```

```python
df2 = df1.groupby('job_title').mean().sort_values('Min experience')
```

```python
df2.columns = ['Min Salary (L) P.A', 'Max Salary (L) P.A', 'Min experience', 'Max experience']
```

```python
df2.head()
```

```python
df2.to_csv('Python Jobs1.csv')
```
